```python
import pandas as pd
irisdata = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", 
                       names=['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class'])  
#print(irisdata.describe(include = 'all'))
X = irisdata.iloc[:,0:4]
y = irisdata.iloc[:,4:5]

#from pandas.plotting import radviz
#radviz(irisdata, 'Class', color=['red','blue','green'])

from sklearn import preprocessing  
le = preprocessing.LabelEncoder()
y = y.apply(le.fit_transform)  #Transform/encode label categorical data to integer

from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20) #Split training and test set

from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler(with_mean=0, with_std=1)  
scaler.fit(X_train) # Fit only to the training data
X_train = scaler.transform(X_train)  # Apply the transformations to the data
X_test = scaler.transform(X_test)  # Apply the transformations to the data
results = []
for i in range(100):
    from sklearn.neural_network import MLPClassifier  
    mlp = MLPClassifier(activation='relu', solver='adam', batch_size='auto', learning_rate_init=0.01, hidden_layer_sizes=(10,10,10,10,10), max_iter=100000)  
    mlp.fit(X_train, y_train.values.ravel())
    predictions = mlp.predict(X_test) 
    print("Training / Test set score: %s / %s percent." %(round(mlp.score(X_train, y_train)*100,1) , round(mlp.score(X_test, y_test)*100,1)))
    results.append(round(mlp.score(X_train, y_train)*100,1))
    from sklearn.metrics import classification_report, confusion_matrix  
    print(confusion_matrix(y_test,predictions))  
    print(classification_report(y_test,predictions))  

import matplotlib.pyplot as plt
plt.plot(results)
plt.xlabel('Runs')
plt.ylabel('Percentage')
plt.title('Training Score %')
plt.show()
```
